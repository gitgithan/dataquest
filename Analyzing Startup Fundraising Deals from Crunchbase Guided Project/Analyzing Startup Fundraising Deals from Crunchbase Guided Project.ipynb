# Count missing values

```python
import pandas as pd

chunks = pd.read_csv('crunchbase-investments.csv',chunksize=5000,encoding = 'iso-8859-1')

mv_counts = []
for chunk in chunks:
    missing = len(chunk) - chunk.count()
    print(missing)    
    mv_counts.append(missing)
```

```python
combined_mv_counts = pd.concat(mv_counts)
combined_mv_counts.groupby(combined_mv_counts.index).sum().sort_values(ascending=False)
```

# Count Sizes of each column

```python
import pandas as pd

chunks = pd.read_csv('crunchbase-investments.csv',chunksize=5000,encoding = 'iso-8859-1')

size_counts = []
for chunk in chunks:
    size_counts.append(chunk.memory_usage(deep=True))
size_counts
```

```python
combined_size_counts=pd.concat(size_counts)
combined_size_counts.groupby(combined_size_counts.index).sum()/1024**2
```

```python
# Key: Column name, Value: List of types
col_types = {}
chunk_iter = pd.read_csv('crunchbase-investments.csv', chunksize=5000, encoding='ISO-8859-1', names=keep_cols)

for chunk in chunk_iter:
    for col in chunk.columns:
        if col not in col_types:
            col_types[col] = [str(chunk.dtypes[col])]
        else:
            col_types[col].append(str(chunk.dtypes[col]))
col_types
```

```python
uniq_col_types = {}
for k,v in col_types.items():
    uniq_col_types[k] = set(col_types[k])
uniq_col_types
```

```python
import sqlite3

conn = sqlite3.connect('crunchbase.db')
chunk_iter = pd.read_csv('crunchbase-investments.csv', chunksize=5000, encoding='ISO-8859-1')

for chunk in chunk_iter:
    chunk.to_sql("investments", conn, if_exists='append', index=False)
```

# Investor that contributed the most across all startups

```python
q = '''
    SELECT investor_name, sum(raised_amount_usd) from investments
    group by investor_name
    order by 2 desc
'''

conn = sqlite3.connect('crunchbase_with.db') 
pd.read_sql(q, conn)
```

# Investors who contributed the most per startup

```python
q = '''
    SELECT company_name,investor_name,max(raised_amount_usd) from investments
    group by company_name
    order by 3 desc
'''

conn = sqlite3.connect('crunchbase_with.db') 
pd.read_sql(q, conn)
```

# Funding round type popularity by count and raised amount

```python
q = '''
    SELECT funding_round_type, count(*), sum(raised_amount_usd) from investments
    group by funding_round_type
    order by 2 desc
'''

conn = sqlite3.connect('crunchbase_with.db') 
pd.read_sql(q, conn)
```
