# 2 threads reduces time from 0.22 to 0.2 seconds, 3 threads 0.19, 5 threads 0.22 again (Times will vary across runs)

```python
import time 
import threading

start = time.time()

content_thread = []

filenames = os.listdir('wiki')

def add_content(files):
    for path in files:
        with open('wiki/{}'.format(path),encoding='utf-8') as f:
            content_thread.append(f.read())
                                  
                       
t1 = threading.Thread(target=add_content, args=(filenames[0:500],)) 
t2 = threading.Thread(target=add_content, args=(filenames[500:1000],))
                           
t1.start()
t2.start()
                           
for thread in [t1, t2]:
    thread.join()
    
        
time.time()-start 
```

# Using Thread Pool

```python
import time 
import concurrent.futures

start = time.time()
content_threadpool = []
filenames = os.listdir('wiki')

def add_content(path):
    with open('wiki/{}'.format(path),encoding='utf-8') as f:
        return f.read()
            
pool = concurrent.futures.ThreadPoolExecutor(max_workers=4)
content_threadpool = list(pool.map(add_content,filenames))

time.time()-start 
```

## Comments: Difference between thread pool and single thread is small, overhead of thread opening/closing may be too large relative to reading time

# Using Threadpool to find tag counts across all html documents

```python
from bs4 import BeautifulSoup

def count_tags(html):
    soup = BeautifulSoup(html, 'html.parser')
    tags = {}
    for tag in soup.find_all():
        if tag.name not in tags:
            tags[tag.name] = 0
        tags[tag.name] += 1
    return tags

start = time.time()
pool = concurrent.futures.ThreadPoolExecutor(max_workers=3)
tags = pool.map(count_tags, content_threadpool)
tags = list(tags)

tag_counts = {}
for tag in tags:
    for k,v in tag.items():
        if k not in tag_counts:
            tag_counts[k] = 0
        tag_counts[k] += v
end = time.time()

print(end - start)

def takeSecond(elem):
    return elem[1]

sorted(tag_counts.items(), key=takeSecond, reverse = True)
```

# Finding top 10 words that at at least 10 Chars long from all html


```python
from bs4 import BeautifulSoup
from collections import Counter
import re

def count_words(html):
    soup = BeautifulSoup(html, 'html.parser')
    words = {}
    text = soup.get_text()
    text = re.sub("\W+", " ", text.lower())
    words = text.split(" ")
    words = [w for w in words if len(w) >= 10]
    return Counter(words).most_common(10)

start = time.time()
pool = concurrent.futures.ThreadPoolExecutor(max_workers=3)
words = pool.map(count_words, content_threadpool)
words = list(words)


word_counts = {}
for wc in words:
    for word, count in wc:
        if word not in word_counts:
            word_counts[word] = 0
        word_counts[word] += 1
end = time.time()

#print(end - start)
word_counts
```
