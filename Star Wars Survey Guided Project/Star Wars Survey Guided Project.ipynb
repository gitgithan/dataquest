```python
import pandas as pd
import matplotlib.pyplot as plt
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.describe(include='all')
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
#plt.show()
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
     np.nan: False
}

for col in star_wars.columns[3:9]:
   
    star_wars[col] = star_wars[col].map(movie_mapping)
    
star_wars
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

### Alternative way to rename besides typing all column names

```python

columns = star_wars.T['Which of the following Star Wars films have you seen? Please select all that apply.':'Unnamed: 8'].index

mapper = {x:f'seen_{idx}' for idx,x in zip(range(1,7),columns)} 
#mapper = {x:f'seen_{idx}' for idx,x in enumerate(columns,start = 1)} #make mapper
star_wars =star_wars.rename(columns=mapper)

star_wars.head()
```

```python
known_words = ['a']
vocab_list = ['a','b','c']
set(word for word in vocab_list if word not in known_words)
```

```python
%matplotlib inline

rank = star_wars.loc[:,'ranking_1':'ranking_6'].mean()
rank.plot.bar()
```

Up to this point, we have cleaned the responses to boolean values and renamed the columns to seen{0} and ranking{0}
On average, the respondents ranked movie 5 the best. Movies 4,5,6 were rating higher than 1,2,3. Maybe 4,5,6 being the older movies had a longer time for discussion and understanding, leading to better appreciation and higher rankings.

```python
views = star_wars.loc[:,'seen_1':'seen_6'].sum()
views.plot.bar()
```

Top 3 most viewed movies are 5,6,1. The older movies are seen by more naturally because they came out earlier. More discussions and cretive observation could occur between more people, leading to higher ratings. There is strong negative correlation between number of people who viewed the movie and ranking number. 

```python
views
```

```python
correlate = pd.concat([rank.reset_index(drop=True),views.reset_index(drop=True)],axis=1)
correlate.corr()
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
ranking_male = males.loc[:,'ranking_1':'ranking_6'].mean()
ranking_male.plot.bar()
plt.show()
views_male = males.loc[:,'seen_1':'seen_6'].sum()
views_male.plot.bar()
plt.show()
ranking_female = females.loc[:,'ranking_1':'ranking_6'].mean()
ranking_female .plot.bar()
plt.show()
views_female = females.loc[:,'seen_1':'seen_6'].sum()
views_female.plot.bar()
```

```python
correlate_male = pd.concat([rank.reset_index(drop=True),views_male.reset_index(drop=True)],axis=1)
correlate_male.corr()
```

```python

correlate_female = pd.concat([ranking_female.reset_index(drop=True),views_female.reset_index(drop=True)],axis=1)
correlate_female.corr()
```

If we split into genders, the negative correlation between views and ranking gets stronger for both genders, especially females at -0.95. Alot more males than females watched 1,2,3 but ranked them alot worse.

```python
fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == "Yes"]
not_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == "No"]
```

```python
fan.loc[:,'ranking_1':'ranking_6'].mean().plot.bar()
plt.show()
fan.loc[:,'seen_1':'seen_6'].sum().plot.bar()
plt.show()
```

```python
not_fan.loc[:,'ranking_1':'ranking_6'].mean().plot.bar()
plt.show()
not_fan.loc[:,'seen_1':'seen_6'].sum().plot.bar()
plt.show()
```

Fans watch all 6 films equally much while non fans watch 1,5,6 alot more. fans rate the original 3 better.
