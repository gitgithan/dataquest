```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
```

```python
recent_grads.iloc[0]
```

```python
pd.options.display.max_columns = 100
recent_grads.describe(include='all')
```

```python
raw_data_count = recent_grads.shape[0]
print(raw_data_count)

recent_grads_clean = recent_grads.dropna()
recent_grads_clean.shape
cleaned_data_count = recent_grads_clean.shape[0]
print(cleaned_data_count)
```

```python
recent_grads.head()
```

```python
ax = recent_grads_clean.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Employed vs. Sample_size')
```

Having more employed gives a larger sample size, possibly because there are more people in the major to sample from, and assuming the survey is sent to all, more people complete the survey

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. Sample_size')
```

Unemployment rate has a larger range at smaller sample sizes. 

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Median vs. Full_time')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Median vs. ShareWomen')
```

Majors with higher share of women make lower income!
